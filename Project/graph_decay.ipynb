```c++
{
    TCanvas *c = new TCanvas();
    TGraph *grafico[2];
    grafico[0] = new TGraph();
    grafico[1] = new TGraph();
    
    TMultiGraph *Multigrafico = new TMultiGraph();
    
    fstream file;
    file.open("population.txt", ios::in);
    
    string line;
    getline(file,line);
    //cout << line << endl;
    while(1)
    {
        float time, count1, count2;
        file >> time >> count1 >> count2;
        
        grafico[0]->SetPoint(grafico[0]->GetN(), time, count1);
        Multigrafico->Add(grafico[0]);
        
        grafico[1]->SetPoint(grafico[1]->GetN(), time, count2);
        Multigrafico->Add(grafico[1]); 
        
        if(file.eof()) break;
    }
    file.close();
    grafico[0]->SetMarkerColor(36);
    grafico[0]->SetMarkerSize(1);
    grafico[0]->SetMarkerStyle(23);
    
    grafico[1]->SetMarkerColor(34);
    grafico[1]->SetMarkerSize(1);
    grafico[1]->SetMarkerStyle(32);
    
    Multigrafico->Draw("ALP");
    Multigrafico->SetTitle("Decay; Time (y) ; Number of atoms ");
    
    c->Draw();
    return c;   
    
}
```
