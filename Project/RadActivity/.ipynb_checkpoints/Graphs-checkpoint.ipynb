```python
%pylab inline
import pandas as pd
```

```python
data = pd.read_csv("population.txt", sep="\s+")
analit = pd.read_csv("Analitica.txt", sep="\s+")
analit.head()
```

```python
fig, ax = subplots(1,2,figsize=(12,4))


ax[0].plot(data['time'],data['PopRad'],'ro',label="Radium")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Population")
ax[0].grid()
ax[0].legend()


ax[1].plot(data['time'],data['PopAct'],'bo',label="Actinium")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Population")
ax[1].grid()
ax[1].legend()

fig.tight_layout()
savefig("Populations")
```

```python
fig, ax = subplots(1,2,figsize=(12,4))

ax[0].plot(analit['t'],analit['NR'],'k',label="Analitic Radium")
ax[0].plot(data['time'],data['PopRad'],'ro',label="Radium")
ax[0].set_xlabel("Time")
ax[0].set_ylabel("Population")
ax[0].grid()
ax[0].legend()

ax[1].plot(analit['t'],analit['NA'],'k',label="Analitic Actinium")
ax[1].plot(data['time'],data['PopAct'],'bo',label="Actinium")
ax[1].set_xlabel("Time")
ax[1].set_ylabel("Population")
ax[1].grid()
ax[1].legend()

fig.tight_layout()
savefig("PopulationsAnalit")
```
